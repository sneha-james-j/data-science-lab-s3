```python
# Data visualization
import matplotlib.pyplot as plt      
x = [1,4,3]  
y = [2,4,1]      
plt.plot(x, y, linewidth=4)  
plt.xlabel('x - axis') 
plt.ylabel('y - axis')   
plt.title('My first graph!')    
plt.show()
```

```python
import matplotlib.pyplot as plt    
x = [1,2,3] 
y = [2,4,1]   
plt.bar(x, y) 
plt.xlabel('x - axis') 
plt.ylabel('y - axis')  
plt.title('My first graph!')  
plt.show()
```

```python
import plotly.express as px
fig = px.line(x=[1, 2, 3], y=[2, 4, 6])
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
print(df)
fig = px.line(df, x="species", y="petal_width")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.bar(df, x="sepal_width", y="sepal_length")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.histogram(df, x="sepal_length", y="petal_width")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x="species", y="petal_width")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x="species", y="petal_width",
                 size="petal_length", color="species")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter_3d(df, x = 'sepal_width',
                    y = 'sepal_length',
                    z = 'petal_width',
                    color = 'species')
fig.show()
```

```python
# import modules
from bokeh.plotting import figure, show
# create figure
p = figure(plot_width = 400, plot_height = 400)
p.square([1, 2, 3, 4, 5], [4, 7, 1, 6, 3], 
         size = 10, color = "red") 
# show the results
show(p)
# import modules
from bokeh.plotting import figure, show
# create figure
p = figure(plot_width = 400, plot_height = 400)
# add a line renderer
p.line([1, 2, 3, 4, 5], [3, 1, 2, 6, 5], 
        line_width = 2, color = "green")
# show the results
show(p)
```
